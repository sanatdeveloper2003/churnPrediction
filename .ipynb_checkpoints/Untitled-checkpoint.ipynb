```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import chi2
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, confusion_matrix
from scipy.stats import ttest_ind

# Load the dataset
dataset = pd.read_csv('dataset_churn.csv')

# Dataset overview
print("Dataset Preview:")
print(dataset.head())

print("\nDataset Information:")
print(dataset.info())

# Clean column names
dataset.columns = dataset.columns.str.strip()

# Drop unnecessary columns
columns_to_drop = ['Unnamed: 0', 'CustomerID']
dataset = dataset.drop(columns=columns_to_drop, errors='ignore')

# Display dataset shape and statistics
print("\nDataset Shape:", dataset.shape)
print("\nDataset Statistics:")
print(dataset.describe())

# Check for missing values
missing_values = dataset.isnull().sum()
print("\nMissing Values by Column:")
print(missing_values)

# Handle missing values
imputer_median = SimpleImputer(strategy='median')
dataset['Age'] = imputer_median.fit_transform(dataset[['Age']]).ravel()

imputer_mode = SimpleImputer(strategy='most_frequent')
for col in ['PaymentMethod', 'Service_Internet']:
    dataset[col] = imputer_mode.fit_transform(dataset[[col]]).ravel()

# Verify no missing values remain
print("\nMissing Values After Imputation:")
print(dataset.isnull().sum())

# Encode categorical features
categorical_features = ['Gender', 'Service_Internet', 'Service_Phone', 'Service_TV',
                        'Contract', 'PaymentMethod', 'StreamingMovies', 'StreamingMusic',
                        'OnlineSecurity', 'TechSupport', 'Churn']

labelencoder = LabelEncoder()
dataset[categorical_features] = dataset[categorical_features].apply(labelencoder.fit_transform)

# Investigate relationships between features and target (Churn)
# Bar charts for categorical features
plt.figure(figsize=(20, 10))
for i, col in enumerate(categorical_features[:-1], 1):
    plt.subplot(3, 4, i)
    sns.countplot(x=col, hue='Churn', data=dataset, palette='coolwarm')
    plt.title(f'{col} vs Churn')
plt.tight_layout()
plt.show()

# Box plots for numerical features
plt.figure(figsize=(15, 10))
numerical_columns = ['Age', 'Tenure', 'MonthlyCharges', 'TotalCharges']
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='Churn', y=col, data=dataset, palette='coolwarm')
    plt.title(f'{col} vs Churn')
plt.tight_layout()
plt.show()

# Correlation heatmap for initial features
plt.figure(figsize=(12, 8))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Initial Features)')
plt.show()

# Hypothesis testing: t-tests for numerical features
print("\nT-Tests for Numerical Features:")
for col in numerical_columns:
    churn_group = dataset[dataset['Churn'] == 1][col]
    no_churn_group = dataset[dataset['Churn'] == 0][col]
    t_stat, p_val = ttest_ind(churn_group, no_churn_group, nan_policy='omit')
    print(f'{col}: t-statistic={t_stat:.4f}, p-value={p_val:.4f}')

# Chi-Square Test for categorical features
print("\nChi-Square Test Results for Categorical Features:")
X_categorical = dataset[categorical_features[:-1]]  # Exclude 'Churn'
Y = dataset['Churn']
chi_scores, p_values = chi2(X_categorical, Y)
chi_square_results = pd.DataFrame({'Feature': X_categorical.columns,
                                   'Chi-Square Score': chi_scores,
                                   'p-value': p_values}).sort_values(by='p-value')
print(chi_square_results)

# Boxplots before handling outliers
plt.figure(figsize=(10, 8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=dataset[col])
    plt.title(f'Boxplot of {col} Before Outlier Handling')
plt.tight_layout()
plt.show()

# Detect and handle outliers via clipping
for col in numerical_columns:
    Q1 = dataset[col].quantile(0.25)
    Q3 = dataset[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    dataset[col] = np.clip(dataset[col], lower_bound, upper_bound)

# Boxplots after handling outliers
plt.figure(figsize=(10, 8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=dataset[col])
    plt.title(f'Boxplot of {col} After Outlier Handling')
plt.tight_layout()
plt.show()

# Visualize numerical feature distributions
plt.figure(figsize=(10, 8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(dataset[col], bins=20, kde=True)
    plt.title(f'{col} Distribution')
plt.tight_layout()
plt.show()
```

```python
# Create new features
dataset['TotalServices'] = (dataset['Service_Internet'] + dataset['Service_Phone'] +
                            dataset['Service_TV'] + dataset['OnlineSecurity'] + dataset['TechSupport'])
dataset['CLV'] = dataset['MonthlyCharges'] * dataset['Tenure']
dataset['AvgMonthlyChargeOverTenure'] = dataset['TotalCharges'] / (dataset['Tenure'] + 1)
dataset['RecentPaymentDrop'] = dataset['MonthlyCharges'] / dataset['AvgMonthlyChargeOverTenure']
dataset['TenurePerService'] = dataset['Tenure'] / (dataset['TotalServices'] + 1)
dataset['TotalChargesRatio'] = dataset['TotalCharges'] / (dataset['MonthlyCharges'] + 1)
dataset['EngagementIndex'] = (0.5 * dataset['Service_Internet'] + 0.3 * dataset['Service_Phone'] + 
                              0.2 * dataset['Service_TV'] + 0.5 * dataset['OnlineSecurity'] + 
                              0.5 * dataset['TechSupport'])

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(dataset.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# Split dataset into train and test sets
X = dataset.drop(columns=['Churn'])
Y = dataset['Churn']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Scale numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Helper function for training and evaluating models
def train_evaluate_plot(model, model_name, X_train, X_test, Y_train, Y_test):
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    print(f'\n{model_name} Accuracy: {accuracy_score(Y_test, Y_pred):.4f}')
    if hasattr(model, "predict_proba"):
        Y_pred_proba = model.predict_proba(X_test)[:, 1]
        print(f'{model_name} ROC-AUC: {roc_auc_score(Y_test, Y_pred_proba):.4f}\n')
    print(classification_report(Y_test, Y_pred))
    sns.heatmap(confusion_matrix(Y_test, Y_pred), annot=True, cmap='coolwarm', fmt='d')
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()

# Train models
models = {
    'Logistic Regression': LogisticRegression(random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42)
}

# Evaluate models
model_results = []
for name, model in models.items():
    print(f"\n{name}:")
    train_evaluate_plot(model, name, X_train_scaled, X_test_scaled, Y_train, Y_test)
    scores = cross_val_score(model, X_train_scaled, Y_train, cv=5, scoring='accuracy')
    model_results.append({
        'Model': name,
        'Mean CV Accuracy': np.mean(scores),
        'Std CV Accuracy': np.std(scores)
    })

# Display model performance comparison
model_comparison = pd.DataFrame(model_results)
print("\nModel Performance Comparison:")
print(model_comparison)

# Hyperparameter tuning for Logistic Regression
param_grid_lr = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['liblinear', 'lbfgs']
}
lr_grid_search = GridSearchCV(LogisticRegression(random_state=42), param_grid_lr, cv=5, scoring='accuracy', n_jobs=-1)
lr_grid_search.fit(X_train_scaled, Y_train)
print(f"\nBest Logistic Regression Params: {lr_grid_search.best_params_}")

# Final evaluation of tuned Logistic Regression
best_lr = lr_grid_search.best_estimator_
print("\nFinal Tuned Logistic Regression:")
train_evaluate_plot(best_lr, "Tuned Logistic Regression", X_train_scaled, X_test_scaled, Y_train, Y_test)

# Hyperparameter tuning for Decision Tree
param_grid_dt = {
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
dt_grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_dt, cv=5, scoring='accuracy', n_jobs=-1)
dt_grid_search.fit(X_train_scaled, Y_train)
print(f"\nBest Decision Tree Params: {dt_grid_search.best_params_}")

# Final evaluation of tuned Decision Tree
best_dt = dt_grid_search.best_estimator_
print("\nFinal Tuned Decision Tree:")
train_evaluate_plot(best_dt, "Tuned Decision Tree", X_train_scaled, X_test_scaled, Y_train, Y_test)

# Hyperparameter tuning for Random Forest
param_grid_rf = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}
rf_grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1)
rf_grid_search.fit(X_train_scaled, Y_train)
print(f"\nBest Random Forest Params: {rf_grid_search.best_params_}")

# Final evaluation of tuned Random Forest
best_rf = rf_grid_search.best_estimator_
print("\nFinal Tuned Random Forest:")
train_evaluate_plot(best_rf, "Tuned Random Forest", X_train_scaled, X_test_scaled, Y_train, Y_test)

# Hyperparameter tuning for Gradient Boosting
param_grid_gb = {
    'n_estimators': [100, 200],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 10]
}
gb_grid_search = GridSearchCV(GradientBoostingClassifier(random_state=42), param_grid_gb, cv=5, scoring='accuracy', n_jobs=-1)
gb_grid_search.fit(X_train_scaled, Y_train)
print(f"\nBest Gradient Boosting Params: {gb_grid_search.best_params_}")

# Final evaluation of tuned Gradient Boosting
best_gb = gb_grid_search.best_estimator_
print("\nFinal Tuned Gradient Boosting:")
train_evaluate_plot(best_gb, "Tuned Gradient Boosting", X_train_scaled, X_test_scaled, Y_train, Y_test)

# Compare performance of tuned models
tuned_model_results = [
    {'Model': 'Logistic Regression', 'Best Params': lr_grid_search.best_params_},
    {'Model': 'Decision Tree', 'Best Params': dt_grid_search.best_params_},
    {'Model': 'Random Forest', 'Best Params': rf_grid_search.best_params_},
    {'Model': 'Gradient Boosting', 'Best Params': gb_grid_search.best_params_}
]
tuned_model_comparison = pd.DataFrame(tuned_model_results)
print("\nTuned Model Comparison:")
print(tuned_model_comparison)
```
